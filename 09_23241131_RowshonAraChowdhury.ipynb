```python

import random
import numpy as np

file_path=('/content/input.txt')
with open(file_path,'r') as file:
    daily_tran=int(file.readline().strip())
    tran_type=[]
    amount=[]

    for i in range (daily_tran):
        t,a =file.readline().strip().split()
        tran_type.append(t)
        a=int(a)
        if t=="d": #as deposits hoile +
            amount.append(a)
        else: #loan hoile -
            amount.append(-a)
print(daily_tran)
print(tran_type)
print(amount)

target=0

#binary strings
def create_population(daily_tran):
    population=[]
    i=0
    while i<20:
        p=np.random.randint(2,size=daily_tran)   #jae lenght 7 and random 0 or 1  [array([0, 1, 0, 0, 1, 1, 0]), array([0, 1, 0, 1, 0, 0, 1]),....]
        population.append(p)
        i+=1
    return population

#check fitness
def cal_fitness(population):
    fitness=[]
    f=0
    while f<len(population):
        total=0
        for index, binary in enumerate(population[f]):
            if binary==1:
                total+=amount[index] #joto num index er value 1 oi given amount from file add
        fitness.append(abs(total))
        f+=1
    return fitness


#kharap ta bad diye dibo
def select_population(fitness):
    temp = []                   #maximum difference removing from the population
    i = 0
    while i < len(fitness):
        temp.append(abs(target - fitness[i]))
        i += 1
    min_fit = temp.index(max(temp)) #index num nise?
    return min_fit


def crossover(x, y):
    cross = []
    point = random.randint(1, daily_tran - 2)  #[0,0,1,0,1,0,1]
    x_first, x_last, y_first, y_last = [], [], [], []

    i = 0
    while i <= point:
        x_first.append(x[i])
        y_first.append(y[i])
        i += 1

    i = point + 1
    while i < daily_tran:
        x_last.append(x[i])
        y_last.append(y[i])
        i += 1

    cross.append(x_first + y_last)
    cross.append(y_first + x_last)
    return cross

def mutation(childx,childy):
    point=random.randint(0, daily_tran - 1) #as binary te ase 1 ke 0 or 0 ke 1 korbo je kono ekta change korbo
    if childx[point] == 1:
        childy[point] = 0
    else:
        childx[point] = 1
    if childy[point] == 1:
        childy[point] = 0
    else:
        childy[point] = 1
    return childx, childy



result = []
for i in range(50):                                       # Loop a fixed number
    population = create_population(daily_tran)  # Generating a new population
    fitness = cal_fitness(population)                # Calculating the fitness of each individual in the population
    min_fit = select_population(fitness)                 # Select the most fit individuals from the population
    population.pop(min_fit)                             # Removing the least fit individual from the population

    solve = False

    for j in range(len(population)):

        x = random.choice(population)                     # Select two individuals at random to crossbreed
        y = random.choice(population)                     # Generating two new children through crossover
        child= crossover(x, y)
        childx, childy=child[0],child[1]
        childx, childy = mutation(childx, childy)
        sum=0
        for index,el in enumerate(childx):
            if el == 1:
                sum+=amount[index]
        if sum == target:
            if childx!=([0]*daily_tran): # sob gula 0 na
                result=childx
                solve = True
                break
        sum=0
        for index,el in enumerate(childy):
            if el == 1:
                sum+=amount[index]
        if sum == target:
            if childy!=([0]*daily_tran):
                result=childy
                solve = True
                break



    if solve == True:

        break
if len(result)>0:
    for i in range(len(result)):
        print(result[i], end='')
else:
    print('-1')
```
